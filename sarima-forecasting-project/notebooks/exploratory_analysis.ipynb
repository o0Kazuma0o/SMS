# Exploratory Data Analysis for Admissions Data

This notebook performs exploratory data analysis (EDA) on the admissions data to visualize and understand trends and patterns in the number of online admission applications.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualisation style
sns.set(style='whitegrid')
```

```python
# Load the processed admissions data
data_path = '../data/processed/admissions_data_processed.csv'
admissions_data = pd.read_csv(data_path)

# Display the first few rows of the dataset
admissions_data.head()
```

```python
# Check for missing values
missing_values = admissions_data.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Visualize the trend of daily admission applications
plt.figure(figsize=(12, 6))
sns.lineplot(data=admissions_data, x='date', y='applications')
plt.title('Daily Admission Applications Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Applications')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Extract month from the date
admissions_data['date'] = pd.to_datetime(admissions_data['date'])
admissions_data['month'] = admissions_data['date'].dt.strftime('%Y-%m')

# Aggregate the total number of applications by month
monthly_applications = admissions_data.groupby('month')['applications'].sum().reset_index()

# Visualize the total number of applications by month
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_applications, x='month', y='applications')
plt.title('Total Admission Applications by Month')
plt.xlabel('Month')
plt.ylabel('Total Applications')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Conclusion

This exploratory analysis provides insights into the trends and patterns of online admission applications. Further analysis can be conducted to understand the factors influencing these trends.
